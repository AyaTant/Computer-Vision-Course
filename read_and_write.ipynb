```python
import cv2
```

```python
image = cv2.imread("./mountain.jpg")
```

```python
# check if image is loaded succressfully
if image is None:
    print(f"Error: Can not find image")
else:
    print("Image loaded successfully")
```

```python
# Display the image
cv2.imshow("Image Window",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_new = image + 30
cv2.imwrite("./output.jpg",image_new)
```
